Este Notebook presentamos un breve EDA de los datasets 

```python
# Importamos las librerias 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cargamos los datos 
items= pd.read_csv('./Data/items.csv')
reviews= pd.read_csv('./Data/Reviews.csv')
games= pd.read_csv('./Data/Steam_games.csv')
users= pd.read_csv('./Data/Users.csv')
```

Hacemos un histograma de los precios para ver como se distribuyen en el dataframe

```python
sns.set_style("white")
plt.figure(figsize=(8, 4))
# Crear el gráfico
sns.histplot(data=games, x='price', color='#48E5C2', bins= 100)

# Agregar título y etiquetas
plt.title('Distribución de Precios de Juegos', fontsize=16, fontweight='bold')
plt.xlabel('Precio', fontsize=14)
plt.ylabel('')

# Ajustar los márgenes
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crea la lista de los top 10
top_generos = games['principal_genre'].value_counts().nlargest(10).index
top_generos = games[games['principal_genre'].isin(top_generos)]
order = top_generos['principal_genre'].value_counts().index
# Crear el gráfico
plt.figure(figsize=(8, 4))  # Ajustar el tamaño de la figura
sns.countplot(data=top_generos, y='principal_genre', color='#6154A0', order=order)
# Agregar título y etiquetas
plt.title('Top 10 de géneros de video juegos', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad', fontsize=14,fontweight='bold')
plt.ylabel('')

# Ajustar los márgenes
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crea la lista de los top 10
top_tags = games['principal_tag'].value_counts().nlargest(10).index
top_tags = games[games['principal_tag'].isin(top_tags)]
order = top_tags['principal_tag'].value_counts().index
# Crear el gráfico
plt.figure(figsize=(8, 4))  # Ajustar el tamaño de la figura
sns.countplot(data=top_tags, y='principal_tag', color='#2176AE', order=order)
# Agregar título y etiquetas
plt.title('Top 10 etiquetas de juegos', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad', fontsize=14,fontweight='bold')
plt.ylabel('')

# Ajustar los márgenes
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crea la lista de los top 10
top_specs = games['principal_spec'].value_counts().nlargest(10).index
top_specs = games[games['principal_spec'].isin(top_specs)]
order = top_specs['principal_spec'].value_counts().index
# Crear el gráfico
plt.figure(figsize=(8, 4))  # Ajustar el tamaño de la figura
sns.countplot(data=top_specs, y='principal_spec', color='#C0B7CC', order= order)
# Agregar título y etiquetas
plt.title('Top 10 especificaciones de juegos', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad', fontsize=14, fontweight='bold')
plt.ylabel('')

# Ajustar los márgenes
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Crea la lista de los top 10
top_years = games['release_year'].value_counts().nlargest(10).index
top_years = games[games['release_year'].isin(top_years)]
order = top_years['release_year'].value_counts().index
sns.set_style("white")
plt.figure(figsize=(8, 4))
# Crear el gráfico
sns.countplot(data=top_years, y='release_year', color='#5F4BB6', order= order)

# Agregar título y etiquetas
plt.title('Top 10 años con más juegos', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad', fontsize=14,)
plt.ylabel('')

# Ajustar los márgenes
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Seleccionar las columnas que contienen 'senti_' en el nombre y calcular sus totales
senti_columns = [col for col in reviews.columns if 'senti_' in col]
totals = reviews[senti_columns].sum()

# Crear un DataFrame para los totales
totals_df = pd.DataFrame({'Sentimiento': totals.index, 'Total': totals.values})
# Cambiar los nombres de las barras

rename_dict = {
    'senti_negativo': 'Negativa',
    'senti_neutral': 'Neutral',
    'senti_positivo': 'Positiva'
}
totals_df['Sentimiento'] = totals_df['Sentimiento'].replace(rename_dict)

# Establecer el estilo del gráfico
sns.set_style("white")

# Crear el gráfico de barras
colors = ['#EC1C1C', '#BFC8AD', '#53FF45']
sns.barplot(x='Sentimiento', y='Total', data=totals_df, palette=colors)
# Agregar título y etiquetas
plt.title('Total de Reviews', fontsize=16, fontweight='bold')
plt.xlabel('Tipo de Reviews', fontsize=14)
plt.ylabel('')

# Ajustar los márgenes y mostrar el gráfico
plt.tight_layout()
plt.show()
```
